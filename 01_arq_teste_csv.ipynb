```python
import pandas as pd
import numpy  as np
import os  # Biblioteca para conversar com as pastas do Windows

# 1. CRIAR A PASTA AUTOMATICAMENTE (A "Mágica")
if not os.path.exists('data'):
    os.makedirs('data')
    print("Pasta 'data' criada com sucesso! 📁")
else:
    print("A pasta 'data' já existe. Seguindo para o salvamento... ✅")

# 2. GERAR OS DADOS (O seu "Big Data" de teste)
df_grande = pd.DataFrame({
    'ID': range(1, 101),
    'Vendedor': np.random.choice(['Jean', 'Ana', 'Carlos', 'Beatriz'], 100),
    'Venda': np.random.uniform(100, 5000, 100).round(2)
})

# 3. SALVAR O ARQUIVO
df_grande.to_csv('data/vendas_grandes.csv', index=False)
print("Arquivo 'vendas_grandes.csv' salvo com sucesso na pasta data!")
```
