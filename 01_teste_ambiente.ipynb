```python
import pandas as pd
import numpy  as np

# Criando um mini-dataset de vendas da Contoso
vendas_contoso = pd.DataFrame({
    'ID_Produto': [101, 102, 103, 104],
    'Qtd_Vendida': [50, 120, 30, 200],
    'Preco_Unitario': [15.50, 25.00, 10.90, 5.00]
})

# Criando uma nova coluna calculada com NumPy
vendas_contoso['Total_Faturamento'] = vendas_contoso['Qtd_Vendida'] * vendas_contoso['Preco_Unitario']

print("--- Relatório de Vendas Profissional ---")
vendas_contoso
```

```python

# Criando um dicionário que simula uma extração do SQL da Contoso
dados_vendas = {
    'Data': pd.to_datetime(['2026-01-01', '2026-01-02', '2026-01-01', '2026-01-03']),
    'Loja': ['São Paulo', 'Rio de Janeiro', 'São Paulo', 'Curitiba'],
    'Produto': ['Notebook', 'Mouse', 'Teclado', 'Notebook'],
    'Valor': [5000, 150, 200, 4800],
    'Imposto': [0.15, 0.10, 0.10, 0.15]
}

df_vendas = pd.DataFrame(dados_vendas)

# Boas Práticas: Criando uma coluna calculada (Lucro Bruto)
# Fórmula: Valor - (Valor * Imposto)
df_vendas['Lucro_Bruto'] = df_vendas['Valor'] * (1 - df_vendas['Imposto'])

print("Primeiras 5 linhas do Relatório Corporativo:")
display(df_vendas.head())
```

```python

# Filtra as linhas e armazena em uma variável
vendas_sp = df_vendas[df_vendas['Loja'] == 'São Paulo']['Valor'].sum() 

# Exibe apenas as linhas filtradas
print(f"O faturamento total de São Paulo foi: R$ {vendas_sp:.2f}")
```

```python
# Agrupando por Loja e somando os valores de tudo
resumo_lojas = df_vendas.groupby('Loja')['Valor'].sum().reset_index()

# Ordenando do maior para o menor (Padrão de relatório de vendas)
resumo_lojas = resumo_lojas.sort_values(by='Valor', ascending=False)

display(resumo_lojas)
```

```python

vendas_rj = df_vendas[df_vendas['Loja'] == 'Rio de Janeiro'] ['Valor'].sum() 


print (f'O faturamento total das lojas do Rio de Janeiro: {vendas_rj:.2f}')
```

```python
# Agrupando por Loja e somando os valores de tudo
resumo_lojas = df_vendas.groupby('Loja')['Valor'].sum().reset_index()

# Ordenando do maior para o menor (Padrão de relatório de vendas)
resumo_lojas = resumo_lojas.sort_values(by='Valor', ascending=False)

display(resumo_lojas)
```

```python
media_valor_notebook = df_vendas[df_vendas['Produto'] == 'Notebook'] ['Valor'].mean()

print (f"Tiket Médio do Notebook é de: {media_valor_notebook:,.2f}")
```

```python
performance_produto = df_vendas.groupby('Produto')['Valor'].sum().reset_index()

performance_produto = performance_produto.sort_values(by='Valor', ascending=False)

display (performance_produto)
```

```python
quantidade_vendas_por_loja = df_vendas.groupby('Loja').count()

# Opção B: O jeito mais rápido do Pandas (direto na coluna)
quantidade_vendas_por_loja = df_vendas['Loja'].value_counts().reset_index()

# Renomeando a coluna 'count' para 'Qtd Ven Lojas'
quantidade_vendas_por_loja = quantidade_vendas_por_loja.rename(columns={'Loja': 'Loja Eletrônico','count': 'Qtd Ven Loja'})

display(quantidade_vendas_por_loja)
```
