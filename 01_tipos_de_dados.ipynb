## Módulo 1, Aula 1: A Natureza dos Dados

Antes de calcularmos qualquer média ou treinarmos um algoritmo de Inteligência Artificial, precisamos entender a "matéria-prima" com a qual estamos lidando. Em Ciência de Dados, classificar a variável corretamente dita as regras do que podemos ou não fazer com ela matematicamente.

### 1. O Conceito

Na estatística, dividimos as variáveis (que serão as colunas do seu DataFrame) em duas grandes famílias:

* **Variáveis Quantitativas (Numéricas):** Representam quantidades mensuráveis.
    * **Contínuas:** Podem assumir qualquer valor dentro de um intervalo, incluindo frações. (Ex: Salário, altura, temperatura, tempo de acesso ao site). Geralmente lidas como `float` no Python.
    * **Discretas:** Assumem apenas valores inteiros, resultantes de contagens. Não existe "meio" valor. (Ex: Número de filhos, quantidade de produtos no carrinho, acessos diários). Lidas como `int`.
* **Variáveis Qualitativas (Categóricas):** Representam qualidades, categorias ou rótulos.
    * **Ordinais:** Existe uma ordem ou hierarquia lógica natural entre as categorias. (Ex: Nível de escolaridade, tamanho de roupa P/M/G, avaliações de 1 a 5 estrelas).
    * **Nominais:** Não há ordem ou hierarquia. São apenas nomes ou rótulos. (Ex: Cores, estado civil, gênero, marcas de carro).

### 2. O Contexto de Mercado

Por que empresas como Ifood ou Nubank se importam com isso?
Primeiro, pela **escolha do modelo**: se você quer prever uma variável Contínua (ex: preço de uma corrida), você usará um modelo de Regressão. Se quer prever uma variável Categórica (ex: o cliente vai cancelar a conta, sim ou não?), você usará um modelo de Classificação. 
Segundo, por **performance**: armazenar variáveis categóricas com o tipo de dado correto no Pandas reduz drasticamente o consumo de memória RAM dos servidores.

### 3. A Prática em Python

No Pandas, o método `.info()` é o seu melhor amigo para o primeiro contato com os dados.

```python
import pandas as pd

# Simulando um banco de dados de clientes
dados = {
    'id_cliente': [101, 102, 103, 104],       # Numérico, mas na verdade é Categórico Nominal
    'idade': [25, 34, 28, 45],                # Quantitativa Discreta
    'salario': [4500.50, 8000.00, 3200.75, 12000.00], # Quantitativa Contínua
    'plano_assinatura': ['Basic', 'Premium', 'Basic', 'Enterprise'] # Qualitativa Ordinal
}

df_clientes = pd.DataFrame(dados)

# Inspecionando os tipos de dados atribuídos automaticamente
print(df_clientes.info())
```

### 4. 💡 Dica de Ouro

**Cuidado com os falsos números!** Códigos postais (CEP), números de telefone, CPFs e IDs de clientes (`id_cliente`) geralmente são compostos por números, mas **são variáveis Qualitativas Nominais**. 

Fazer cálculos matemáticos com eles não faz sentido. Qual é a média dos números de telefone de São Paulo? Não significa nada. No mercado, deixamos essas colunas como texto (`object` ou `string`) para evitar que algoritmos tentem somar ou calcular médias dessas colunas acidentalmente.

````python
import pandas as pd

# Simulando um banco de dados de clientes 

dados = {
    'id_produto': [1001, 1002, 1003, 1004],     # Variável Qualitativa Nominal 
    'categoria_produto': ['Eletrônico', 'Roupas', 'Alimentos', 'Bebidas'], # Variável Qualitativa Nominal 
    'quantidade_estoque': [100, 300, 500, 850], # Variável quantitativa discreta
    'peso_kg': [1.5, 1.0, 3.0, 0.5],            # Variável Quantitativa Continua 
    'tamanho': ['Pequeno', 'Médio', 'Grande', 'Extra Grande']              # Variável Qualitativa Ordinal 
}

df_produtos = pd.DataFrame(dados)

# Conversão explícita para String

print('--- Antes ---')

df_produtos.info()

df_produtos['id_produto'] = df_produtos['id_produto'].astype(str) # Conversão para string 

print('--- Depois ---')

(df_produtos.info())
````
